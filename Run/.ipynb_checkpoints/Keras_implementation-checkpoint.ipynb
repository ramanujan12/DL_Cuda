```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from load_mnist import load_data_wrapper
import matplotlib.pyplot as plt 
```

```python
inputs = keras.Input(shape=(784,))
dense = layers.Dense(64, activation="sigmoid")
x = dense(inputs)
x = layers.Dense(64, activation="sigmoid")(x)
outputs = layers.Dense(10,activation="softmax")(x)
model = keras.Model(inputs=inputs, outputs=outputs, name="mnist_model")
model.summary()
```

```python
model.compile(
    loss=keras.losses.CategoricalCrossentropy(),
    optimizer=keras.optimizers.SGD(learning_rate=.1),
    metrics=[keras.metrics.CategoricalCrossentropy(),"accuracy"],
)
```

```python
train,val,test=load_data_wrapper()
batchsize=64
batches=[]
n_batches=int(train[0].shape[0]/batchsize)
for batch in range(n_batches):
    batches+=[((train[0][batch*batchsize:(batch+1)*batchsize]),\
                (train[1][batch*batchsize:(batch+1)*batchsize]),\
                (val[0][:batchsize]),\
                (val[1][:batchsize]))]



# (x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()

# x_train = x_train.reshape(60000, 784).astype("float32") / 255
# x_test = x_test.reshape(10000, 784).astype("float32") / 255

# y_train_one_hot=np.array([np.eye(1,10,x).reshape(-1) for x in y_train])
# y_test_one_hot=([np.eye(1,10,x).reshape(-1) for x in y_test])
```

```python


history = model.fit(train[0], train[1], batch_size=64, epochs=2, validation_split=0.2)
# history = model.fit(x_train, y_train_one_hot, batch_size=64, epochs=2, validation_split=0.2)

# test_scores = model.evaluate(x_test, y_test, verbose=2)
# print("Test loss:", test_scores[0])
# print("Test accuracy:", test_scores[1])
```

```python
hist=[]
for batch in batches:
    hist+=[model.train_on_batch(batch[0],batch[1])]
```

```python
hist_array=np.array(hist)
plt.plot(np.arange(len(hist)),hist_array[:,2])
hist_array[-10:,2].mean()
```

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from load_mnist import load_data_wrapper
import matplotlib.pyplot as plt 
```

```python
train,val,test=load_data_wrapper()


inputs = keras.Input(shape=(784,))
outputs = layers.Dense(10,activation="sigmoid")(inputs)
model_ol = keras.Model(inputs=inputs, outputs=outputs, name="mnist_model")
model_ol.summary()

model_ol.compile(
    loss=keras.losses.MeanSquaredError(),
    optimizer=keras.optimizers.SGD(learning_rate=.1),
    metrics=[keras.metrics.CategoricalCrossentropy(),"accuracy"],
)
```

```python
history = model_ol.fit(test[0], test[1], batch_size=1000, epochs=20, validation_split=0.2)
```
